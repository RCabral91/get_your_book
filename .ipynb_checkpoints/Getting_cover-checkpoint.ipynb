```python
import pandas as pd

df = pd.read_csv("/home/rafael/Área de Trabalho/CUROS IRONHACK/SEMANA 10/get_your_book/data/raw/processed/books_final_version.csv")

# 1) Remover a coluna antiga de capa
if "cover_image_uri" in df.columns:
    df.drop(columns=["cover_image_uri"], inplace=True)

# 2) Função para criar capa baseada no ISBN
def isbn_to_cover(isbn):
    if pd.isna(isbn):
        return "https://via.placeholder.com/150x220?text=No+Cover"
    isbn = str(isbn).strip().replace("-", "")
    return f"https://covers.openlibrary.org/b/isbn/{isbn}-L.jpg?default=false"

# 3) Criar nova coluna
df["cover_image_uri"] = df["isbn13"].apply(isbn_to_cover)

# 4) Salvar o novo CSV
df.to_csv("books_final_version_cleaned.csv", index=False)
print("✅ Capas recriadas com sucesso!")
```
