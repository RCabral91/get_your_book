```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
import urllib.parse
from urllib.parse import urljoin
```

```python
url = "https://books.toscrape.com/"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser") 

books = soup.find_all("article", class_="product_pod")
data = []

rating_dict = {"One": 1, "Two": 2, "Three": 3, "Four": 4, "Five": 5}

for book in books:
    title = book.h3.a["title"]
    price = book.find("p", class_="price_color").text[1:]
    rating = rating_dict[book.find("p")["class"][1]]
    
    raw_link = book.h3.a["href"]
    link = urljoin("https://books.toscrape.com/", raw_link)

    r = requests.get(link)
    bs = BeautifulSoup(r.text, "html.parser")

    breadcrumb = bs.find("ul", class_="breadcrumb")
    if breadcrumb is None:
        continue
        
    category = breadcrumb.find_all("li")[2].text.strip()

    availability = bs.find("p", class_="instock availability").text.strip()

    desc_tag = bs.find("article", class_="product_page").find("p", recursive=False)
    description = desc_tag.text.strip() if desc_tag else ""
    
    img = bs.find("div", class_="item active").img["src"]
    img = urljoin(link, img)

    data.append({
        "title": title,
        "category": category,
        "price": price,
        "rating": rating,
        "availability": availability,
        "image_url": img,
        "link": link,
        "description": description,
    })

print(len(data))
```
