```python
import pandas as pd
import requests
from concurrent.futures import ThreadPoolExecutor

CSV_PATH = "/home/rafael/√Årea de Trabalho/CUROS IRONHACK/SEMANA 10/get_your_book/data/raw/processed/books_final_version_cleaned.csv"

df = pd.read_csv(CSV_PATH)

# ------------------------
# 1. Detectar capas faltantes/ruins
# ------------------------
def is_missing_cover(url):
    if pd.isna(url) or url == "":
        return True
    if not isinstance(url, str):
        return True
    if not url.startswith("http"):
        return True
    return False


df["needs_cover"] = df["cover_image_uri"].apply(is_missing_cover)

missing_df = df[df["needs_cover"]].copy()
print(f"üîç Capas faltantes: {missing_df.shape[0]}")

# ------------------------
# 2. Fun√ß√£o para buscar capa
# ------------------------
def fetch_cover(isbn):
    isbn = str(isbn).strip()
    if not isbn or isbn.lower() == "nan":
        return None

    # Open Library API
    url = f"https://covers.openlibrary.org/b/isbn/{isbn}-L.jpg?default=false"
    try:
        resp = requests.head(url, timeout=3)
        if resp.status_code == 200:
            return url
    except:
        pass

    return None


# ------------------------
# 3. Buscar em paralelo apenas para ISBNs faltantes
# ------------------------
with ThreadPoolExecutor(max_workers=20) as executor:
    new_covers = list(executor.map(fetch_cover, missing_df["isbn13"]))

missing_df["new_cover"] = new_covers

# ------------------------
# 4. Combinar com o dataset original
# ------------------------
for idx, row in missing_df.iterrows():
    if row["new_cover"]:
        df.at[idx, "cover_image_uri"] = row["new_cover"]
    else:
        df.at[idx, "cover_image_uri"] = "https://via.placeholder.com/150x220?text=No+Cover"

# ------------------------
# 5. Salvar resultado
# ------------------------
SAVE_PATH = "/home/rafael/√Årea de Trabalho/CUROS IRONHACK/SEMANA 10/get_your_book/data/raw/processed/books_final_version_cleaned.csv"

df.to_csv(SAVE_PATH, index=False)
print(f"‚úÖ Arquivo atualizado salvo em:\n{SAVE_PATH}")
```

```python
import pandas as pd
import requests
from pathlib import Path

# ===== PATHS =====
INPUT_FILE = "/home/rafael/√Årea de Trabalho/CUROS IRONHACK/SEMANA 10/get_your_book/data/raw/processed/books_final_version_cleaned.csv"
OUTPUT_FILE = "/home/rafael/√Årea de Trabalho/CUROS IRONHACK/SEMANA 10/get_your_book/data/raw/processed/books_final_version_cleaned.csv"

df = pd.read_csv(INPUT_FILE)

# ===== Fun√ß√µes auxiliares =====

def valid_image(url):
    """Retorna True se imagem tem tamanho > 5KB, evitando as imagens 1x1."""
    try:
        r = requests.get(url, timeout=4)
        if r.status_code == 200 and len(r.content) > 5000:
            return True
        return False
    except:
        return False

def get_openlibrary_cover(isbn):
    isbn = str(isbn).replace("-", "").strip()
    url = f"https://covers.openlibrary.org/b/isbn/{isbn}-L.jpg?default=false"
    return url if valid_image(url) else None

def get_google_cover(isbn):
    isbn = str(isbn).replace("-", "").strip()
    api_url = f"https://www.googleapis.com/books/v1/volumes?q=isbn:{isbn}"
    try:
        r = requests.get(api_url, timeout=4).json()
        items = r.get("items", [])
        if not items:
            return None
        links = items[0]["volumeInfo"].get("imageLinks", {})
        # thumbnail usually works
        thumb = links.get("thumbnail")
        if thumb:
            return thumb.replace("&zoom=1", "&zoom=2")  # better quality
    except:
        return None
    return None

PLACEHOLDER = "https://via.placeholder.com/150x220?text=No+Cover"

# ===== Processamento =====
final_covers = []

for isbn in df["isbn13"]:
    if pd.isna(isbn):
        final_covers.append(PLACEHOLDER)
        continue
    
    # 1. tenta OpenLibrary
    cover = get_openlibrary_cover(isbn)
    if cover:
        final_covers.append(cover)
        continue
    
    # 2. tenta Google Books
    cover = get_google_cover(isbn)
    if cover:
        final_covers.append(cover)
        continue
    
    # 3. fallback ‚Üí placeholder
    final_covers.append(PLACEHOLDER)

# adiciona ao dataframe
df["cover_image_uri"] = final_covers

# salva
df.to_csv(OUTPUT_FILE, index=False)
print("‚úÖ Capas atualizadas com sucesso!")
print(f"üìÅ Salvo em: {OUTPUT_FILE}")
```
