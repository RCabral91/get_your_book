```python
# =========================
# 0) Imports & setup
# =========================
import os, joblib, numpy as np, pandas as pd
from pathlib import Path
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import normalize
from scipy.sparse import save_npz

# --- caminhos (ajuste se quiser) ---
DATA_PATH   = Path("/home/rafael/√Årea de Trabalho/CUROS IRONHACK/SEMANA 10/get_your_book/data/raw/processed")
INPUT_CSV   = DATA_PATH / "books_final.csv"
OUTPUT_DIR  = DATA_PATH / "ml_outputs"
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

# =========================
# 1) Carregar dados
# =========================
df = pd.read_csv(INPUT_CSV, engine="python")
print("shape:", df.shape)
print("colunas:", df.columns.tolist()[:20])

# =========================
# 2) Texto base para NLP
# =========================
for col in ["title", "author", "description"]:
    if col not in df.columns:
        df[col] = ""
df["title"]       = df["title"].fillna("").astype(str)
df["author"]      = df["author"].fillna("").astype(str)
df["description"] = df["description"].fillna("").astype(str)

df["text"] = df["title"] + ". " + df["author"] + ". " + df["description"]

USE_STOPWORDS = True

# =========================
# 3) Vetoriza√ß√£o TF-IDF
# =========================
vectorizer = TfidfVectorizer(
    stop_words=("english" if USE_STOPWORDS else None),
    ngram_range=(1,2),
    max_df=0.9,
    min_df=3,
    max_features=20000
)
X = vectorizer.fit_transform(df["text"])
X_norm = normalize(X, norm="l2", copy=True)

print("‚úÖ TF-IDF shape:", X.shape)

# =========================
# 4) Busca de K (dataset completo)
# =========================
print("\nüîç Calculando Silhouette Score para v√°rios K (com dataset completo)...")

Ks = list(range(4, 16))
scores = {}

for k in Ks:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X_norm)
    sil = silhouette_score(X_norm, labels)
    scores[k] = sil
    print(f"K={k:2d} ‚Üí silhouette={sil:.4f}")

best_k = max(scores, key=scores.get)
print(f"\n‚≠ê Melhor K = {best_k} ‚Üí Silhouette = {scores[best_k]:.4f}")

# =========================
# 5) Modelo final
# =========================
kmeans = KMeans(n_clusters=best_k, random_state=42, n_init=10)
labels = kmeans.fit_predict(X_norm)
df["cluster"] = labels

# =========================
# 6) Termos representativos
# =========================
feature_names = np.array(vectorizer.get_feature_names_out())
order_centroids = np.argsort(kmeans.cluster_centers_, axis=1)[:, ::-1]

TOP_N = 12
cluster_terms = []
for c in range(best_k):
    terms = feature_names[order_centroids[c, :TOP_N]].tolist()
    cluster_terms.append(", ".join(terms))

short_labels = [", ".join(t.split(", ")[:3]) for t in cluster_terms]
cluster_df = pd.DataFrame({
    "cluster": list(range(best_k)),
    "top_terms": cluster_terms,
    "label": short_labels
})
print("\n=== Cluster Overview ===")
display(cluster_df)

# mapear labels curtos
label_map = dict(zip(cluster_df["cluster"], cluster_df["label"]))
df["cluster_label"] = df["cluster"].map(label_map)

# =========================
# 7) Recomendador (cosine)
# =========================
from sklearn.metrics.pairwise import linear_kernel

def recommend_by_title(query_title, top_n=10):
    idx = df.index[df["title"].str.lower() == str(query_title).lower()]
    if len(idx) == 0:
        idx = df.index[df["title"].str.lower().str.startswith(str(query_title).lower())]
    if len(idx) == 0:
        return pd.DataFrame(columns=["title","author","score","cluster_label"])
    i = idx[0]
    sims = linear_kernel(X_norm[i], X_norm).ravel()
    sims[i] = -1
    top_idx = sims.argsort()[::-1][:top_n]
    out = df.loc[top_idx, ["title","author","cluster_label"]].copy()
    out["score"] = sims[top_idx]
    return out[["title","author","score","cluster_label"]]

# exemplo: recommend_by_title("frankenstein", top_n=5)

# =========================
# 8) Salvar tudo
# =========================
df_out_csv = OUTPUT_DIR / "books_clustered.csv"
df.to_csv(df_out_csv, index=False, encoding="utf-8")
print("\nüíæ CSV salvo em:", df_out_csv)

joblib.dump(vectorizer, OUTPUT_DIR / "tfidf_vectorizer.joblib")
joblib.dump(kmeans,     OUTPUT_DIR / "kmeans_model.joblib")
save_npz(OUTPUT_DIR / "tfidf_matrix_norm.npz", X_norm)
pd.Series(scores, name="silhouette").to_csv(OUTPUT_DIR / "k_silhouette_scores.csv")

print("üß† Modelos e matrizes salvos em:", OUTPUT_DIR)
```

```python
# ======================================================
# ETAPA 7 - SISTEMA DE RECOMENDA√á√ÉO BASEADO EM CONTE√öDO
# ======================================================

from sklearn.metrics.pairwise import cosine_similarity

# 1Ô∏è‚É£ Carregar artefatos (caso precise rodar isolado)
#    Use os nomes corretos se voc√™ salvou em outro caminho
import joblib
import pandas as pd
from pathlib import Path

DATA_PATH = Path("/home/rafael/√Årea de Trabalho/CUROS IRONHACK/SEMANA 10/get_your_book/data/raw/processed/ml_outputs")

# 1Ô∏è‚É£ Carregar artefatos salvos
df = pd.read_csv(DATA_PATH / "books_clustered.csv", encoding="utf-8")
tfidf = joblib.load(DATA_PATH / "tfidf_vectorizer.joblib")
X_tfidf = load_npz(DATA_PATH / "tfidf_matrix_norm.npz")

# 2Ô∏è‚É£ Garantir colunas limpas
df["title"] = df["title"].astype(str).str.strip().str.lower()
df["author"] = df["author"].astype(str).str.strip()
df["description"] = df["description"].fillna("")

# 3Ô∏è‚É£ Fun√ß√£o de recomenda√ß√£o
def recommend_books(title_query, top_n=5):
    """
    Retorna livros mais similares com base no t√≠tulo/descri√ß√£o/autor.
    """
    title_query = title_query.lower().strip()

    # Verifica se o t√≠tulo existe no dataset
    matches = df[df["title"].str.contains(title_query, case=False, na=False)]
    if matches.empty:
        print(f"‚ùå Nenhum livro encontrado contendo: '{title_query}'")
        return None

    # Pega o primeiro match
    idx = matches.index[0]
    book_vec = X_tfidf[idx]

    # Similaridade com todos
    sims = cosine_similarity(book_vec, X_tfidf).flatten()
    top_idx = sims.argsort()[::-1][1:top_n+1]  # exclui o pr√≥prio livro

    recs = df.loc[top_idx, ["title", "author", "genre", "avg_rating", "cluster_label"]]
    print(f"\nüîç Livro base: {df.loc[idx, 'title'].title()} ({df.loc[idx, 'author']})")
    print("üìö Recomenda√ß√µes:")
    display(recs.reset_index(drop=True))

# 4Ô∏è‚É£ Testes
recommend_books("Frankenstein", top_n=5)
recommend_books("Pride and Prejudice", top_n=5)
```

```python
import pandas as pd
from pathlib import Path

# =====================================================
# üîß Caminhos
# =====================================================
DATA_PATH = Path("/home/rafael/√Årea de Trabalho/CUROS IRONHACK/SEMANA 10/get_your_book/data/raw")
MAIN_PATH = DATA_PATH / "processed/books_final.csv"
GOODREADS_PATH = DATA_PATH / "Goodreads_books_with_genres.csv"
DETAILS_PATH = DATA_PATH / "Book_Details.csv"

# =====================================================
# 1Ô∏è‚É£ Carregar datasets
# =====================================================
df_main = pd.read_csv(MAIN_PATH)
df_good = pd.read_csv(GOODREADS_PATH)
df_det = pd.read_csv(DETAILS_PATH)

print("‚úÖ Shapes:")
print(f"Main: {df_main.shape}, Goodreads: {df_good.shape}, Details: {df_det.shape}")

# =====================================================
# 2Ô∏è‚É£ Padronizar colunas para merge
# =====================================================
# Normaliza nomes
df_main.columns = df_main.columns.str.strip().str.lower()
df_good.columns = df_good.columns.str.strip().str.lower()
df_det.columns = df_det.columns.str.strip().str.lower()

# Assegura que isbn13 √© string
for df in [df_main, df_good, df_det]:
    if "isbn13" in df.columns:
        df["isbn13"] = df["isbn13"].astype(str).str.replace(r"\.0$", "", regex=True).str.strip()

# =====================================================
# 3Ô∏è‚É£ Selecionar apenas colunas relevantes para merge
# =====================================================
df_good_sub = df_good[[
    "isbn13", "average_rating", "publisher", "publication_date", "genres"
]].copy()

df_det_sub = df_det[[
    "book_id", "cover_image_uri", "genres", "average_rating", "num_ratings", 
    "num_reviews", "author", "publication_info"
]].copy()

# =====================================================
# 4Ô∏è‚É£ Merge por ISBN13 com Goodreads
# =====================================================
df_merged = df_main.merge(df_good_sub, on="isbn13", how="left", suffixes=("", "_good"))

# =====================================================
# 5Ô∏è‚É£ Merge por autor e/ou t√≠tulo com Book_Details (para pegar capa e g√™neros extras)
# =====================================================
# Normaliza para compara√ß√£o textual
for df in [df_merged, df_det_sub]:
    for col in ["author"]:
        if col in df.columns:
            df[col] = df[col].astype(str).str.lower().str.strip()

df_merged = df_merged.merge(
    df_det_sub,
    on="author", how="left", suffixes=("", "_det")
)

# =====================================================
# 6Ô∏è‚É£ Combinar g√™neros (Goodreads + Book_Details)
# =====================================================
def combine_genres(row):
    g1 = str(row.get("genres", "")).strip()
    g2 = str(row.get("genres_det", "")).strip()
    if g1 and g2:
        return ", ".join(sorted(set((g1 + ", " + g2).split(", "))))
    return g1 or g2 or None

df_merged["genres_combined"] = df_merged.apply(combine_genres, axis=1)

# =====================================================
# 7Ô∏è‚É£ Consolidar colunas finais √∫teis (corrigido)
# =====================================================
df_final = df_merged.copy()

# Fun√ß√£o auxiliar para preencher colunas com seguran√ßa
def safe_fill(target_col, source_col):
    if target_col not in df_final.columns:
        df_final[target_col] = None
    if source_col in df_final.columns:
        df_final[target_col] = df_final[target_col].fillna(df_final[source_col])

# Preenche as colunas principais
safe_fill("genre", "genres_combined")
safe_fill("avg_rating", "average_rating")
safe_fill("publisher", "publisher_good")
safe_fill("published_date", "publication_date")

# Adiciona a capa
if "cover_image_uri" not in df_final.columns:
    df_final["cover_image_uri"] = None
df_final["cover_image_uri"] = df_final["cover_image_uri"].fillna(
    "https://via.placeholder.com/100x150?text=No+Cover"
)

# =====================================================
# 8Ô∏è‚É£ Limpeza de colunas tempor√°rias
# =====================================================
cols_drop = [
    "genres_det", "genres_good", "genres_combined", "average_rating", 
    "average_rating_good", "publication_date", "publisher_good"
]
df_final.drop(columns=[c for c in cols_drop if c in df_final.columns], inplace=True, errors="ignore")

# =====================================================
# 9Ô∏è‚É£ Exportar resultado
# =====================================================
OUTPUT_PATH = DATA_PATH / "processed/books_final_enriched.csv"
df_final.to_csv(OUTPUT_PATH, index=False, encoding="utf-8")
print(f"\n‚úÖ Dataset enriquecido salvo em: {OUTPUT_PATH}")
print("Dimens√µes finais:", df_final.shape)
print("\nValores ausentes (principais colunas):")
print(df_final[["genre", "avg_rating", "publisher", "published_date", "cover_image_uri"]].isna().sum())
```

```python
# =====================================================
# üîé 1) Verificar duplicatas por ISBN13
# =====================================================
print("\n=== Verifica√ß√£o de duplicatas por ISBN13 ===")

# Conta quantas vezes cada ISBN aparece
isbn_counts = df_final["isbn13"].value_counts()
duplicates = isbn_counts[isbn_counts > 1]

print(f"üîç ISBNs duplicados: {len(duplicates)}")
if len(duplicates) > 0:
    print("Exemplo de duplicados:")
    dup_rows = df_final[df_final["isbn13"].isin(duplicates.index)]
    display(dup_rows.head(10)[["isbn13", "title", "author", "genre", "avg_rating"]])

# =====================================================
# üßπ 2) Remover duplicatas mantendo o registro mais completo
# =====================================================
# Conta quantos valores n√£o nulos h√° em cada linha
df_final["non_null_count"] = df_final.notna().sum(axis=1)

# Mant√©m apenas o mais completo por ISBN13
df_final_clean = (
    df_final.sort_values("non_null_count", ascending=False)
    .drop_duplicates(subset=["isbn13"], keep="first")
    .drop(columns=["non_null_count"], errors="ignore")
    .reset_index(drop=True)
)

print(f"\n‚úÖ Ap√≥s limpeza: {df_final_clean.shape[0]} registros √∫nicos (antes: {df_final.shape[0]})")




```
